# Evaluation Metrics

We wish to use Confusion Matrix, specifically the values of Recall in order to limit false negatives. Recall is defined as “TruePositives / (TruePositives + FalseNegatives).” The reason why we are using Recall is due to the fact that it can be extremely harmful if someone has a heart disease but the model predicts that they don't – a false negative – which is measured by Recall. Specifically, Recall will be a value between 0 and 1, with 0 being the worst and 1 being perfect Recall. This model is able to tolerate false positives (Measured by Precision) because that would mean that a patient not diagnosed with heart disease is tested positive, all that leads to for the patient is further medical analysis. If we evaluate our model with the F1 metric, which equally gives weight to Recall and Precision, it would put some weight away from prioritizing the maximization of the Recall value. However, we will still be also looking at other metrics that involve FalseNegatives such as False Negative Rate, False Omission Rate, and F1 score.

```python
####### BENCHMARK #######
# Building a model using KNeighborsClassifier 
def evaluate_model(model, x_test, y_test):
    from sklearn import metrics

    # Predict Test Data 
    y_pred = model.predict(x_test)

    # Calculate accuracy, precision, recall, f1-score, and kappa score
    acc = metrics.accuracy_score(y_test, y_pred)
    rec = metrics.recall_score(y_test, y_pred)
    f1 = metrics.f1_score(y_test, y_pred)

    # Calculate area under curve (AUC)
    y_pred_proba = model.predict_proba(x_test)[::,1]

    # Display confussion matrix
    cm = metrics.confusion_matrix(y_test, y_pred)

    return {'acc': acc, 'rec': rec, 'f1': f1, 'cm': cm}
knn = KNeighborsClassifier(n_neighbors = 5)

knn.fit(X_train, y_train)


# Evaluate Model
knn_eval = evaluate_model(knn, X_test, y_test)

# Print result
print('Accuracy:', knn_eval['acc'])
print('Recall:', knn_eval['rec'])
print('F1 Score:', knn_eval['f1'])
print('Confusion Matrix:\n', knn_eval['cm'])
```

### Interpreting the result

For most of the classifiers, we were getting the exact same accuracy for both training and testing, which tells us that they are converging into a maximum accuracy based on the dataset. However, for Random Forests Classification and Support Vector Machine Classifier, we are actually seeing higher testing accuracy than the rest.

* K Nearest-Neighbors: 0.80
* Random Forest: 0.82

* Logistic Regression: 0.80

As you can see, Random Forest has higher recall than the other models. This means that Random Forest on average does better in predicting correctly that someone has heart disease based on this metric.

When comparing our recall to our baseline KNN model, our results were able to improve upon both values that had a recall of 0.72.

Additionally, looking at the best scores performed by the models gave us the following results.

* K Nearest-Neighbors: 0.78
* Random Forest: 0.81
* SVM: 0.83
* Logistic Regression: 0.79

From this, it is apparent that SVM has the highest overall performance achieved compared to the other models. This means that SVM on average does better in predicting correctly that someone has heart disease based on this metric. 

However, basically all of our models except for Logistic Regression had to use very small hyperparameter searches. From the remaining three though, SVM stood out as the longest to run and thus had its parameters limited the most.

Specifically, SVM had only two regularization parameter options due to our constrained time and belief. Our model ended up with the smaller value of the two as more ideal. This means that our model used wider margins allowing for more misclassification errors but possibly less overfitting that would be present in higher values of C where the margins would be narrower. We also looked at a small and large value of gamma and our model selected a smaller value meaning that our decision boundary was most likely less complex but also hopefully less prone to overfitting with a better fit to the training data.

For Random Forest our max depth took upon a larger max depth from our offered values meaning that our data was more accurate in fitting the training data which could potentially lead to a more overfit model. Our model also chose a Gini criterion meaning that our data was split based on the impurity measure of misclassification. Using this one instead of the other common one Entropy, means that our model was most likely less sensitive to the class imbalance. For the minimum samples to split on was the smallest value we provided meaning that our model might be less prone to overfitting and better at generalization. The n estimators chose the usual default of 100 meaning that we had a high but not highest value of trees present suggesting a relatively sufficient accuracy that could be higher. 

Our KNN model used a K number of neighbors that was the highest from the values we searched over meaning our model is most likely more robust to noise which is a benefit considering the sheer volume of our data. This though is at the cost of our model being less accurate at its predictions. We also saw that a power parameter used the value of 2 of Euclidean distance meaning that our model was probably using more geometric distance fittings to the data points. Our p value is still small though and thus places less weight on bigger distances between data points meaning that we might have underfit. Finally the weights were uniform meaning that all the points in the neighborhood were weighed equally but possibly less robust to noise though meaning more prone to overfit.

Finally, our Logistic regression searched over many values for its regularization term and ended using a rather smaller value of C meaning that we have stronger regularization and hopefully thus less overfitting. The model also ended up with an L1 penalty and its corresponding solver Saga for our high dimensional data. Furthermore SAGA’s SGD is better at converging faster. Our L1 penalty will punish outliers more which is probably more ideal in our high dimensional data with many features and is thus hopefully less prone to overfitting.  

In summation, our Logistic Regression model is probably the most prone to overfitting, noise, and best at regularization from all of our models. Though our SVM similarly seemed less prone to overfitting but had much less parameters to search over. Similarly our KNN was prone to underfitting due to being cautious of overfits. Our Random Forest though showed more possibility of overfitting and having poor generalization.


### Limitations

There are multiple limitations in both our data and the approach we took. For one, the data is coming from one country and while the results are large, there is always a possibility that they overrepresent, and thus underrepresent some groups. Additionally, our data was unbalanced, and while we took approaches to correct for that, it could also still have affected our results leading to an overrepresentation of people without heart disease.

Furthermore, we had to structure our models in a way that would allow us to run them in a timely manner in order to complete this project. Because we were using Google Colab, we were only offered limited amounts of computational resources, and we did not have infinite time. We also had a relatively larger dataset that limited our ability to perform a more extensive search of possible parameters. As such, our hyperparameters are most likely not as extensive as they could be. This is especially true for our SVM model, as well as KNN and Random Forests to a smaller degree, as they simply would not run with larger hyperparameters.

We felt like it would be better to explore more hyperparameters if we had more time and computational resources. For example, we would have added different types of regularization for the other models. For logistic regression we were only testing L1 and L2 regularization, but we would have tried elasticnet with alpha value intervals of 0.1. Another case were SVMs, where we would have tested out different kernels such as RBF or polynomial. Unfortunately, due to the highly computationally complex nature of SVMs, we were only able to make do with the linear kernel. We also could only use very few regularization parameters to search through so that is another parameter we would have wanted to explore further, specifically higher values of C to see if narrower margins would have been more ideal. 

This is also true for our KNN and Random Forests. Our KNN for example could have used more hyperparameters such as more neighbors to values to test as the one chosen as best was the highest value we gave meaning we could have used even higher values to be more robust to noise. Similarly, our Random Forest could have used a much more extensive maximum depth hyperparameter with a wider range of values to search through to improve our accuracy with the best number of trees and avoid overfit as the max depths values were very sparse.

Also, the variables we chose to be crucial might have or have not been and thus have added more inaccuracy to our model. Similarly, we might have not included variables that were actually very important in predicting heart disease. 

We feel like more data would definitely change the nature of the problem because we're only taking a subset of risk factors associated with heart disease. Specifically, more data could make our models more accurate with better crucial variables. 

For example, heart disease could be part of a genetic problem. We would need a variable in this case for each individual to state whether or not their genes are faulty. Heart disease could also be based on specific factors in your diet. Maybe people with heart disease have a deficiency of calcium or vitamin B6. We will need variables about the patient in order to include that in our predictions for whether or not someone has heart disease.

This also highlights another limitation in our data. As our data was collected through surveys done over the phone, they are prone to people not reporting as accurately as one might expect physician acquired results. A survey also limited the results that were collected as physical health had to be measured through a more abstract application such as the number of days where it was reported bad.
